## Membuat tabel yang sesuai

Tabel yang akan dibentuk terdiri atas kolom-kolom berikut:

Tabel tersebut merupakan hasil penggabungan (merge) antara data **masterdesa** dan data dari **Data Dinas Sosial**.  
Proses penggabungan dilakukan berdasarkan kolom identitas wilayah administratif, seperti:

- `kode_provinsi`
- `nama_provinsi`
- `kabupaten`
- `kecamatan`
- `kelurahan`

Beberapa kolom juga dilakukan penyesuaian nama agar struktur tabel akhir konsisten dan dapat menampung informasi dari kedua sumber data.


```python
# Mengubah nama kolom agar sesuai dengan nama tabel yang diinginkan
data1 = data1.rename(columns={
    'bps_desa_kelurahan':'bps_nama_desa_kelurahan',
    'kemendagri_desa_kelurahan':'kemendagri_nama_desa_kelurahan'
})
```

```python
# Membersihkan data sebelum melakukan merge
def clean(df, kolom):
    for k in kolom:
      # Menghapus spasi berlebih dan mengubah isi kolom menjadi huruf kapital
        df[k] = df[k].astype(str).str.strip().str.upper()

# Membersihkan kolom-kolom pada data1 (masterdesa) agar memiliki format yang seragam
clean(data1, ['nama_provinsi', 'nama_kabupaten_kota', 'bps_nama_kecamatan', 'bps_nama_desa_kelurahan'])
# Membersihkan kolom-kolom pada data2 (Data Dinas Sosial) agar memiliki format yang seragam
clean(data2, ['nama_provinsi', 'kabupaten', 'kecamatan', 'kelurahan'])
# Mengubah format penulisan kabupaten pada data2
data2['kabupaten'] = data2['kabupaten'].str.replace('KAB. ', 'KABUPATEN ', regex=False)
# Menyamakan nama beberapa kolom pada data1 agar sesuai dengan kolom pada data2
data1['kabupaten'] = data1['nama_kabupaten_kota']
data1['kecamatan'] = data1['bps_nama_kecamatan']
data1['kelurahan'] = data1['bps_nama_desa_kelurahan']
```

```python
# Melakukan penggabungan data berdasarkan 'kode_provinsi', 'nama_provinsi','kabupaten', 'kecamatan', 'kelurahan'
data = pd.merge(
    data2,
    data1,
    how='left',
    on=['kode_provinsi', 'nama_provinsi','kabupaten', 'kecamatan', 'kelurahan']
)
```

```python
# Menampilkan data yang sudah digabungkan
data.head()
```

```python
# Melihat nama-nama kolom pada data yang sudah digabungkan
data.columns.tolist()
```

```python
# Mengurutkan kolom berdasarkan ketentuan pada tabel yang diinginkan
data_urut=['id','kode_provinsi',
           'nama_provinsi','kode_kabupaten_kota','kabupaten','nama_kabupaten_kota',
           'bps_kode_kecamatan','kecamatan','bps_nama_kecamatan','bps_kode_desa_kelurahan',
           'kelurahan','bps_nama_desa_kelurahan','kemendagri_kode_kecamatan',
           'kemendagri_nama_kecamatan','kemendagri_kode_desa_kelurahan',
           'kemendagri_nama_desa_kelurahan','periode_update','nik','nama',
           'alamat','kategori','jumlah','satuan','tahun']
```

```python
# Memastikan hanya kolom yang ada dalam DataFrame yang digunakan
data_urut = [col for col in data_urut if col in data.columns]
```

```python
# Menyusun ulang kolom DataFrame sesuai urutan tersebut
data = data[data_urut]
```

## Data Cleansing

```python
# Melihat informasi terkait banyak baris yang ada pada masing-masing kolom, hal ini penting untuk mengetahui perbedaan setelah dilakukan penanganan data yang hilang
data.info()
```

```python
# Memeriksa nilai yang hilang pada masing-masing kolom
data.isna().sum()
```

 Dilihat dari hasil pemeriksaan pada data, terdapat sejumlah nilai yang hilang (`null`). Salah satu cara umum untuk mengatasi nilai yang hilang adalah dengan menghapus baris yang memiliki nilai kosong. Langkah ini dilakukan untuk menjaga keakuratan data, terutama jika kolom yang kosong memegang peranan penting dalam analisis.

Selain menghapus, cara lain yang sering digunakan untuk menangani nilai hilang adalah mengisi nilai kosong (imputasi) dengan nilai tertentu, seperti rata-rata, median, modus, atau nilai tetap tertentu. Namun, pemilihan metode tergantung pada konteks data dan tujuan analisis.

```python
# Menangani nilai yang hilang
data = data.dropna()
```

```python
# Memeriksa kembali apakah masih terdapat nilai yang hilang pada data
data.isna().sum()
```

```python
# Melihat banyaknya baris pada masing-masing kolom setelah dilakukan penanganan nilai yang hilang
data.info()
```
